```python

def getActorsForMovie(title, year):
    search = tmdb.Search()

    search.movie(query=title, primary_release_year=year)

    if search.results:
        movieActors = {}
        for result in search.results:
            movieTitle = result.get('original_title')

            movieID = result.get('id')
            cast = tmdb.Movies(movieID).credits().get('cast')
            listOfActors = []
            for actor in cast: 
                listOfActors.append(actor.get('name'))
            
            movieActors[movieTitle] = listOfActors
        return movieActors
    return "No movie found"

getActorsForMovie('Gladiator', 2000)
```
